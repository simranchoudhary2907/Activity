```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([3,7,9,13,15])

n = len(x)

sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x*y)
sum_x2 = np.sum(x**2)

m = (n*sum_xy - sum_x*sum_y) / (n*sum_x2 - sum_x**2)
b = (sum_y - m*sum_x) / n

print("Slope (m):", m)
print("Intercept (b):", b)
```

```python
import matplotlib.pyplot as plt

y_pred = m*x + b

plt.scatter(x, y, label="Actual Data")
plt.plot(x, y_pred, label="Best Fit Line")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
print("Predicted values:", y_pred)
```

```python
error = y - y_pred
sse = np.sum(error**2)

print("Sum of Squared Error:", sse)
```

```python
x_new = x
y_new = m*x_new + b

plt.scatter(x_new, y_new, label="Predicted Points")
plt.plot(x_new, y_new, color='red')

plt.title("f_new(x) = mx + b")
plt.show()
```

```python
import pandas as pd

# create random data
x = np.random.randint(40,50,100)
y = np.random.randint(40,50,100)

df = pd.DataFrame({"x": x, "y": y})
df.to_csv("data.csv", index=False)

df.head()
```

```python
# read file
df = pd.read_csv("data.csv")
df
```

```python
# Step 5: Calculate m and b (Least Squares)
n = len(df)

sum_x = np.sum(df["x"])
sum_y = np.sum(df["y"])
sum_xy = np.sum(df["x"] * df["y"])
sum_x2 = np.sum(df["x"]**2)

m = (n*sum_xy - sum_x*sum_y) / (n*sum_x2 - sum_x**2)
b = (sum_y - m*sum_x) / n
```

```python
# Step 7: Plot (Point 3 - Regression Line)
y_pred = m * df["x"] + b
plt.scatter(df["x"], df["y"], label="Actual Data")
plt.plot(df["x"], y_pred, label="Best Fit Line")

plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Step 8: Plot (Point 5 - f_new(x))
plt.scatter(df["x"], y_pred, color='red', label="Predicted Points")
plt.plot(df["x"], y_pred)

plt.title("f_new(x) = mx + b")
plt.legend()
plt.show()
```
