```python
df = pd.DataFrame({
"Product": ["A","B"],
"Category": ["X","Y"],
"Price": [100,200],
"Stock": [10,20]
})

df
```

```python
df["Discount"] = np.random.randint(5,20,size=len(df))
df
```

```python
df["Final Price"] = df["Price"] - (df["Price"] * df["Discount"]/100)
df
```

```python
df = df.drop("Stock", axis=1)
df
```

```python
df = pd.DataFrame({
"Student":["A","B","C"],
"Algorithm":[85,40,90],
"Data Structure":[88,45,92],
"English":[80,30,85]
})

df
```

```python
# Above 80 in all
df1 = df[(df.drop('Student', axis=1) > 80).all(axis=1)]
df1
```

```python
df2 = df[(df.drop('Student',axis=1)<50).any(axis=1)]
df2
```

```python
df["Total"] = df[["Algorithm","Data Structure","English"]].sum(axis=1)
df
```

```python
df3 = df[df["Total"] >= df["Total"].quantile(0.95)]
df3
```
